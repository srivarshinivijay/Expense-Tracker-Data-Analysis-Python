```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Data = [
    ["2026-01-02", "Food", 250, "Lunch"],
    ["2026-01-03", "Travel", 120, "Metro"],
    ["2026-01-05", "Shopping", 1500, "Clothes"],
    ["2026-01-08", "Entertainment", 400, "Movie"],
    ["2026-01-10", "Bills", 2000, "Electricity Bill"],
    ["2026-01-15", "Food", 300, "Dinner"],
    ["2026-01-18", "Travel", 200, "Cab"],
    ["2026-01-22", "Health", 800, "Medicines"],
    ["2026-02-01", "Food", 280, "Breakfast"],
    ["2026-02-03", "Shopping", 2200, "Shoes"],
    ["2026-02-05", "Bills", 1500, "Internet"],
    ["2026-02-07", "Entertainment", 600, "Concert"],
    ["2026-02-10", "Travel", 180, "Bus"],
    ["2026-02-14", "Food", 450, "Valentine Dinner"],
    ["2026-02-18", "Health", 1200, "Doctor Visit"],
    ["2026-03-02", "Food", 320, "Lunch"],
    ["2026-03-05", "Shopping", 1800, "Accessories"],
    ["2026-03-08", "Travel", 250, "Train"],
    ["2026-03-12", "Bills", 2100, "Rent"],
    ["2026-03-15", "Entertainment", 500, "Streaming Subscription"]
]

df = pd.DataFrame(Data, columns=["Date", "Category", "Amount", "Description"])
df
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df.info()
```

```python
total_spending = df["Amount"].sum()
print("Total Spending: ₹", total_spending)
```

```python
category_summary = df.groupby("Category")["Amount"].sum()
category_summary
```

```python
category_summary.plot(kind="pie", autopct="%1.1f%%")
plt.title("Expense Distribution by Category")
plt.ylabel("")
plt.show()
```

```python
category_summary.plot(kind="bar")
plt.title("Category-wise Spending")
plt.xlabel("Category")
plt.ylabel("Total Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_summary = df.groupby(df["Date"].dt.to_period("M"))["Amount"].sum()
monthly_summary
```

```python
monthly_summary.plot(kind="bar")
plt.title("Monthly Spending Trend")
plt.xlabel("Month")
plt.ylabel("Total Amount")
plt.show()
```

```python
monthly_summary.plot(kind="line", marker="o")
plt.title("Monthly Spending Trend")
plt.xlabel("Month")
plt.ylabel("Total Amount")
plt.show()
```

```python
## Insights

- Bills category has the highest total spending (~33%).
- Shopping is the second highest expense category (~32%).
- February shows higher spending compared to January.
- Travel contributes the least to total expenses (~4%).
```
